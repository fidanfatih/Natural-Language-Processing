```python
from bs4 import BeautifulSoup as bs
```

```python
with open('5807689.xml', 'r') as f:
    data = f.read()

data[:5]
```

```python
soup = bs(data, 'html.parser')

soup
```

```python
www = soup.find_all('w')
```

```python
[w.text.strip().replace('\n', ' ') for w in www]
```

```python
sss = soup.find_all('s')
```

```python
[s.text.strip().replace('\n', ' ') for s in sss]
```
